```python
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = glob.glob("./ext_frames/*.png")
print(len(img_path))
```

```python
dummy_img = cv2.imread(img_path[0],cv2.IMREAD_GRAYSCALE)
plt.imshow(dummy_img)
```

```python
#dummy_bin =  cv2.adaptiveThreshold(dummy_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
#            cv2.THRESH_BINARY,151,-1)

kernel = np.ones((5,5),np.uint8)
_, dummy_bin = cv2.threshold(255-dummy_img,190,255,cv2.THRESH_BINARY)
dummy_bin = cv2.erode(dummy_bin,kernel,iterations = 2)
plt.imshow(dummy_bin)
```

```python
for idx, tmp_path in enumerate(img_path):
    tmp_save = tmp_path+"_bin.png" 
    print(tmp_path)
    tmp_img = cv2.imread(tmp_path,cv2.IMREAD_GRAYSCALE)
    _, tmp_bin = cv2.threshold(255-tmp_img,190,255,cv2.THRESH_BINARY)
    tmp_bin = cv2.resize(tmp_bin,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_NEAREST)
    tmp_bin = cv2.erode(tmp_bin,kernel,iterations = 2)
    tmp_bin = cv2.dilate(tmp_bin,kernel,iterations = 2)
    tmp_save = tmp_save.replace("ext_frames","bin_frames")
    print(tmp_save)
    cv2.imwrite(tmp_save,tmp_bin)
```
